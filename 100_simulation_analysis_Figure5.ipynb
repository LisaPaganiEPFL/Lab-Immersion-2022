# Extracting saved dynamics

Useful functions to extract the dynamic

```python
def ratios(fix, x, deme):
    
    #Number of mutants in deme in the first x steps of the dynamics
    deme_mutant=fix[:x,deme,0]

    #Number of wild-types in deme in the first x steps of the dynamics
    deme_WT=fix[:x,deme,1]

    #Ratio of mutant in deme in the first x steps of the dynamics
    deme_ratioM=deme_mutant/(deme_WT+deme_mutant)
    
    return deme_ratioM


def extract(path, x, deme):
    
    deme_dynamics = np.zeros([100,x])
    
    for i in range(100):
        path_sub = path + '/fix_' + str(i+1) + '.npz'
        fix = np.load(path_sub)['arr_0']
        deme_dynamics[i] = ratios(fix,x,deme)
    
    return deme_dynamics
```

## Asymmetric well-mixed configuration

High migration rate

```python
#Path to the asymmetric well-mixed configuration with high migration rate
path = "Asymmetric/Well-mixed/High"

x=40

#Ratio of mutant in demes in the first x steps of the dynamics
center_ratioM = extract(path, x, 0)
leaf1_ratioM = extract(path, x, 1)
leaf2_ratioM = extract(path, x, 2)
leaf3_ratioM = extract(path, x, 3)

center_ratioM_mean = np.mean(center_ratioM, axis= 0)
leaf1_ratioM_mean = np.mean(leaf1_ratioM, axis= 0)
leaf2_ratioM_mean = np.mean(leaf2_ratioM, axis= 0)
leaf3_ratioM_mean = np.mean(leaf3_ratioM, axis= 0)

center_ratioM_std = np.std(center_ratioM, axis= 0,ddof=1)
leaf1_ratioM_std = np.std(leaf1_ratioM, axis= 0,ddof=1)
leaf2_ratioM_std = np.std(leaf2_ratioM, axis= 0,ddof=1)
leaf3_ratioM_std = np.std(leaf3_ratioM, axis= 0,ddof=1)

#Evolution of the ratio of mutants in the center and in the leaves
plt.errorbar(np.arange(0,x,1),center_ratioM_mean, center_ratioM_std, color='darkviolet', label='Center')
plt.errorbar(np.arange(0,x,1),leaf1_ratioM_mean, leaf1_ratioM_std, color='turquoise', label='Leaf 1 (mutant)')
plt.errorbar(np.arange(0,x,1),leaf2_ratioM_mean, leaf2_ratioM_std, color='orange', label='Leaf 2')
plt.errorbar(np.arange(0,x,1),leaf3_ratioM_mean, leaf3_ratioM_std, color='orange', label='Leaf 3')
plt.ylabel('Mutant ratio')
plt.xlabel('Dilution step')
plt.title('Asymmetric, high migration, well-mixed')
plt.legend()
plt.savefig('h-wm.pdf',format='pdf')
plt.show()
```

```python
def p(a,array,clr,x=False):
    
    if(x):
        a.plot([1,4,7,9],[array[1],array[4],array[7],array[9]],'-ok',color=clr)
    else:
        a.plot([1,4,7],[array[1],array[4],array[7]],'-ok',color=clr)
    
    return

def pl(a,array1,array2,array3,array4,clr1,clr2,clr3,clr4,x=False):
    
    p(a,array1,clr1,x)
    p(a,array2,clr2,x)
    p(a,array3,clr3,x)
    p(a,array4,clr4,x)
    
    return
```
