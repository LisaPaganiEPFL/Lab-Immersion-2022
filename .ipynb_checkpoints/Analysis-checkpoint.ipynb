```python
from numba_code import *
import matplotlib.pyplot as plt
plt.style.use('default')
```

# Extracting saved dynamics

## Asymmetric OUT > IN configuration

High migration rate

Folder 'Asymmetric/Out_in/High' contains fixation trajectories up till the 1000th step, labeled as 'fix', and extinction trajectories labeled as 'ex'. 

```python
#Upload one trajectory, gives an numpy array
fix1 = np.load("Asymmetric/Out_in/High/fix_1.npz")['arr_0']
```

```python
#Number of mutants in deme 0 (i.e. center) in the first 50 steps of the dynamics
x=40
center_mutant=fix1[:x,0,0]

#Number of wild-types in deme 0 (i.e. center) in the first 500 steps of the dynamics
center_WT=fix1[:x,0,1]

#Ratio of mutant in deme 0 (center) in the first 500 steps of the dynamics
center_ratioM=center_mutant/(center_WT+center_mutant)
```

```python
#Ratio of mutant in leaf 1, that is initially mutant & wild-type
leaf1_mutant=fix1[:x,1,0]
leaf1_WT=fix1[:x,1,1]
leaf1_ratioM=leaf1_mutant/(leaf1_WT+leaf1_mutant)
```

```python
#Ratio of mutant in leaf 2, that is initially all wild-type
leaf2_mutant=fix1[:x,2,0]
leaf2_WT=fix1[:x,2,1]
leaf2_ratioM=leaf2_mutant/(leaf2_WT+leaf2_mutant)
```

```python
#Ratio of mutant in leaf 3, that is initially all wild-type
leaf3_mutant=fix1[:x,3,0]
leaf3_WT=fix1[:x,3,1]
leaf3_ratioM=leaf3_mutant/(leaf3_WT+leaf3_mutant)
```

```python
#Evolution of the ratio of mutants in the center and in the leaves
plt.plot(np.arange(0,x,1),center_ratioM, color='darkviolet', label='Center')
plt.plot(np.arange(0,x,1),leaf1_ratioM, color='turquoise', label='Leaf 1 (mutant)')
plt.plot(np.arange(0,x,1),leaf2_ratioM, color='orange', label='Leaf 2')
plt.plot(np.arange(0,x,1),leaf3_ratioM, color='orange', label='Leaf 3')
plt.ylabel('Mutant ratio')
plt.xlabel('Dilution step')
plt.title('Asymmetric, high migration, OUT>IN')
plt.legend()
plt.show()
```

```python
high_out_center_ratioM = center_ratioM
high_out_leaf1_ratioM = leaf1_ratioM
high_out_leaf2_ratioM = leaf2_ratioM
high_out_leaf3_ratioM = leaf3_ratioM
```

```python
low_out_center_ratioM = center_ratioM
low_out_leaf1_ratioM = leaf1_ratioM
low_out_leaf2_ratioM = leaf2_ratioM
low_out_leaf3_ratioM = leaf3_ratioM
```

```python
#Evolution of the ratio of mutants in the center and in the leaves
plt.plot(np.arange(0,x,1),low_out_center_ratioM, color='darkviolet', label='Center low')
plt.plot(np.arange(0,x,1),low_out_leaf1_ratioM, color='turquoise', label='Leaf 1 (mutant) low')
plt.plot(np.arange(0,x,1),low_out_leaf2_ratioM, color='orange', label='Leaf 2 low')
#plt.plot(np.arange(0,100,1),leaf3_ratioM, color='orange', label='Leaf 3')
plt.plot(np.arange(0,x,1),high_out_center_ratioM, color='orchid', label='Center high')
plt.plot(np.arange(0,x,1),high_out_leaf1_ratioM, color='aqua', label='Leaf 1 (mutant) high')
plt.plot(np.arange(0,x,1),high_out_leaf2_ratioM, color='gold', label='Leaf 2 high')
#plt.plot(np.arange(0,100,1),high_leaf3_ratioM, color='gold', label='Leaf 3')
plt.ylabel('Mutant ratio')
plt.xlabel('Dilution step')
plt.title('Asymmetric, high and low migration, OUT>IN')
plt.legend()
plt.show()
```
